## NLP Project 7: Hotel Review Analysis
## Problem Statement
A hotel management company collects customer reviews from booking websites and travel platforms to
understand guest experiences related to room quality, staff behavior, cleanliness, food service, and
amenities. Analyzing these reviews manually is challenging because of the large amount of feedback
received from customers. The company wants to apply NLP techniques to analyze guest reviews,
measure customer satisfaction, and identify areas that require service improvement.
## Tasks
1. Collect hotel review data.
2. Preprocess guest reviews.
3. Perform sentiment analysis.
4. Identify service-related issues.
5. Generate customer satisfaction report.

## Step 1: Import Libraries

```python
import pandas as pd
import re
import nltk

nltk.download('stopwords')
nltk.download('vader_lexicon')

from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer

from sklearn.feature_extraction.text import CountVectorizer
```

```python
reviews = [
    "The room was clean and comfortable.",
    "The staff were very friendly and helpful.",
    "The food was delicious.",
    "Excellent service and beautiful rooms.",
    "Very satisfied with my stay.",
    "The room was dirty.",
    "Food quality was poor.",
    "The staff were rude.",
    "The bathroom was not clean.",
    "Very disappointed with the service.",
    "Amazing breakfast and clean rooms.",
    "The hotel location was perfect.",
    "WiFi was very slow.",
    "Air conditioning was not working.",
    "The swimming pool was clean.",
    "Room service was excellent.",
    "The bed was uncomfortable.",
    "Parking was convenient.",
    "The check-in process was quick.",
    "Overall a wonderful experience."
]

df = pd.DataFrame({
    "Review": reviews
})

df.head()
```

## Step 2: Preprocess

```python
stop_words = set(stopwords.words("english"))

def clean_text(text):

    text = text.lower()

    text = re.sub(r'[^a-zA-Z ]','',text)

    words = text.split()

    words = [word for word in words if word not in stop_words]

    return " ".join(words)

df["Cleaned_Review"] = df["Review"].apply(clean_text)

df.head()
```

## Step 3: Perform Sentiment Analysis

```python
sia = SentimentIntensityAnalyzer()

def sentiment(text):

    score = sia.polarity_scores(text)["compound"]

    if score >= 0.05:
        return "Positive"

    elif score <= -0.05:
        return "Negative"

    else:
        return "Neutral"

df["Sentiment"] = df["Cleaned_Review"].apply(sentiment)

df
```

```python
vectorizer = CountVectorizer(stop_words="english")

X = vectorizer.fit_transform(df["Cleaned_Review"])

words = vectorizer.get_feature_names_out()

counts = X.sum(axis=0).A1

issue_df = pd.DataFrame({
    "Word": words,
    "Count": counts
})

issue_df = issue_df.sort_values(by="Count", ascending=False)

issue_df.head(10)
```

## Step 4: Generate Customer Satisfaction Report

```python
report = df["Sentiment"].value_counts()

print("Customer Satisfaction Report")
print(report)
```
